```python
numbers3d_1 = open('numbers3d_1.txt')
numbers3d_2 = open('numbers3d_2.txt')
results = open('results.txt')

nums1x, nums1y = [], []
nums2x, nums2y = [], []
interx, intery = [], []

for line in numbers3d_1:
    nums = list(map(float, line.split(' ')))
    nums1x.append(nums[0])
    nums1y.append(nums[1])
for line in numbers3d_2:
    nums = list(map(float, line.split(' ')))
    nums2x.append(nums[0])
    nums2y.append(nums[1])
for line in results:
    nums = list(map(float, line.split(',')))
    interx.append(nums[0])
    intery.append(nums[1])

    
    
plt.figure(figsize=(20, 20))
plt.scatter(nums1x, nums1y, s=0.5, c='r')
plt.scatter(nums2x, nums2y, s=0.5, c='b')
plt.scatter(interx, intery, s=20, c='g')
```
